```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

```python
(x_train,y_train),(x_test,y_test)=tf.keras.datasets.mnist.load_data()
```

here we have a dataset of 60000 images where every image is an numpy  array of 28*28.

```python
x_train.shape
```

in y_train we have the information of  output of those x_train data.

```python
y_train
```

```python
x_test.shape
```

this is just an example to show how the images are looking .

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[0])
```

```python
x_train[0]
```

we make thev range of values between 0 and 1 so that the output  converges . this will be done by dividing x_train and x_test by 255.

```python
x_train=x_train/255
x_test=x_test/255
```

now we can see that every value is betweeen 0 and 1.

```python
x_train[0]
```

we have to convert tihs 2-D  array info into 784 inputs data with 1 row next to the other row , basically from 2-D to 1-D .

```python
model=Sequential()
model.add(Flatten(input_shape=(28,28)))# this is to change input data from 2-D data to 1-D .
model.add(Dense(128,activation="relu"))# thiss is the first hiddden layer , here we dont have to give input_dim because we have used Flatten in the first add .
model.add(Dense(32,activation='relu'))
model.add(Dense(10,activation='softmax'))# since this is a multiclass classification we will use softmax as a activation function.
```

there are total 100480 trainable parameters as 128*784(weights)+ 128(biases) in 1st hidden layer also ... 1290 trainable parameters in the 2nd hidden layer as 128*10(weights) + 10(biases) .

intotal we have 101770 trainable paarameters.

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
history=model.fit(x_train,y_train,epochs=25,validation_split=0.2)
```

```python
y_prob=model.predict(x_test)
```

```python
y_prob
```

```python
y_pred=y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```
